```python
import pandas as pd
import requests
```

#Import Data

```python
url = 'https://www.pro-football-reference.com/years/2023/fantasy.htm' # HTTP request reflector
table_data = pd.read_html(url)
df = table_data[0]
df.head()
```

#Clean data

```python
#pass columns list as df columns
my_columns = ['rk','player','tm','fantpos','age','g','gs','passing_cmp','passing_att','passing_yds','passing_td','passing_int','rushing_att','rushing_yds','rushing_y_a','rushing_td',
              'receiving_tgt','receiving_rec','receiving_yds','receiving_y_r','receiving_td','fmb','fl','td','2pm','2pp','fantpt','ppr','dkpt','fdpt','vbd','posrank','ovrank']

df.columns = my_columns

df = (
    df.loc[df['rk']!='Rk'] #exclude rows with column names
    .fillna(0)
)

df[df.columns[4:33]] = df[df.columns[4:33]].apply(pd.to_numeric)
df.head()
```
